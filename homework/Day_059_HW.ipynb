### 用 digits (3個數字即可 , dataset.load_digits(n_class=3)) 資料嘗試跑 PCA


```python
from sklearn import datasets

digits = datasets.load_digits(n_class=3)
X = digits.data
y = digits.target
```

載入 相關套件 並 執行 PCA 實驗 ...

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


from sklearn import decomposition

```

```python
np.random.seed(1)

pca = decomposition.PCA(n_components = 3)

pca.fit(X)
X1 = pca.transform(X)

fig = plt.figure(1,figsize = (8,6))
#plt.clf()
ax = Axes3D(fig,rect = [0,0,0.95,1],elev = 48, azim = 134)
#plt.cla()
for name, label in [('digit 0', 0), ('digit 1', 1), ('digit 2', 2)]:
    ax.text3D(X1[y==label,0].mean(),
              X1[y == label, 1].mean() + 1.5,
              X1[y == label, 2].mean(), name,
              horizontalalignment='center',
              bbox=dict(alpha=.5, edgecolor='w', facecolor='w'))
y = np.choose(y,[1,2,0]).astype(np.float)
ax.scatter(X1[:,0],X1[:,1],X1[:,2],c = y,cmap = plt.cm.nipy_spectral, edgecolor = 'k')
ax.w_xaxis.set_ticklabels([])
ax.w_yaxis.set_ticklabels([])
ax.w_zaxis.set_ticklabels([])
ax.set_title('Digit Data After PCA process', size = 14)
```
