```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns # 另一個繪圖-樣式套件
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
matrix = (np.random.random((10,10))- 0.5)*2

plt.figure(figsize=(10,10))
heatmap = sns.heatmap(matrix,cmap = plt.cm.Spectral,annot = True,vmin=-1,vmax=1,square = True)
plt.title('Random Matrix Correlation Heatmap');
```

```python
nrow = 1000
ncol = 3
matrix = (np.random.random((nrow,ncol))-0.5)*2
indice = np.random.choice([0,1,2],size = nrow)
plot_data = pd.DataFrame(matrix,indice).reset_index()

grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
# 上半部為 scatter
grid.map_upper(plt.scatter, alpha = 0.2)

# 對角線畫 histogram
grid.map_diag(sns.distplot)

# 下半部放 density plot
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.suptitle('Random Matrix Pairs Plot', size = 32, y = 1.05)
plt.show()
```

```python
matrix = 5+3*np.random.randn(nrow * ncol).reshape((nrow, ncol))#更改N(mu, sigma^2)為 N(5,3)
indice = np.random.choice([0,1,2],size = nrow)
plot_data = pd.DataFrame(matrix,indice).reset_index()

grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False, hue = 'index', vars = [x for x in list(plot_data.columns) if x != 'index'])
# 上半部為 scatter
grid.map_upper(plt.scatter, alpha = 0.2)

# 對角線畫 histogram
grid.map_diag(sns.distplot)

# 下半部放 density plot
grid.map_lower(sns.kdeplot, cmap = plt.cm.OrRd_r)

plt.suptitle('Random Matrix Pairs Plot', size = 32, y = 1.05)
plt.show()
```
